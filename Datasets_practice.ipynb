```python
ndf['type'].unique()
```

```python
#replacing null values
ndf['endYear']=ndf.loc[(ndf['type'].isin(['tvSeries','tvSpecial', 'tvMiniSeries','tvEpisode']))]['endYear'].fillna('Still Showing')
ndf['endYear']=ndf.loc[(ndf['type'].isin(['movie','short', 'video','videoGame'])),'endYear'].fillna(ndf.startYear)
ndf['endYear']=ndf.loc[(ndf['type']=='No Type'),'endYear'].fillna('Unknown')
```

```python
movies=ndf.loc[(ndf['type'].isin(['tvSeries','tvSpecial', 'tvMiniSeries','tvEpisode']))].shape[0]
tv_shows=ndf.loc[(ndf['type'].isin(['movie','short', 'video','videoGame']))].shape[0]
```

```python
import matplotlib.pyplot as plt
```
