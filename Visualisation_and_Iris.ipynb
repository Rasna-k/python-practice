# Iris

```python
import pandas as pd
```

### 1.Loading data

```python
namelist = ["SepalLength" , "SepalWidth","PetalLength","PetalWidth" , "species"]
idf=pd.read_csv('https://raw.githubusercontent.com/sarwansingh/Python/master/ClassExamples/data/iris.csv',header=None,names =namelist)
```

### 2.Preprocessing data

```python
idf.head()
```

```python
idf['SepalLength'].count()
```

```python
idf['SepalLength'].value_counts()
```

```python
idf.info()
```

```python
idf.describe()
```

```python
idf.groupby('species')['SepalLength'].count()
```

```python
idf['species'].value_counts()
```

```python
# finding out those entries whose petal width is greater than the average petal width
idf[idf['PetalWidth']>idf['PetalWidth'].mean()]
```

```python
idf[idf['PetalWidth']>idf['PetalWidth'].mean()]['PetalWidth'].count()
```

```python
idf[idf['PetalWidth']>idf['PetalWidth'].mean()]['PetalWidth'].value_counts()
```

**find those records who have petal length greater than those of Iris-Setova**

```python
b=idf.loc[idf.species=='Iris-setosa']['PetalLength'].max()
```

```python
idf.loc[idf.PetalLength>b]
```
